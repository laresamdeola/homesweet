```python
capture = cv2.VideoCapture(0)
while capture.isOpened():
    ret, frame = capture.read()
    results = model(frame)

    cv2.imshow('YOLO', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
capture.release()
cv2.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join('yolov5', 'data', 'images')
labels = ['safe', 'not safe']
number_images = 5
```

```python
print(IMAGES_PATH)
```

```python
capture = cv2.VideoCapture(0)

for label in labels:
    print(f"Collecting images for {label}")
    time.sleep(5)

    for num in range(number_images):
        print(f"Collecting images for {label}, image number {num}")
        ret, frame = capture.read()
        image_name = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(image_name, frame)
        cv2.imshow('Image Collection', frame)
        time.sleep(2)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
capture.release()
cv2.destroyAllWindows()
```
